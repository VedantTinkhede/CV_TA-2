```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

# Load image in color and convert to numpy array
img_color = Image.open(r"C:\Users\ACER\Desktop\CV_TA-2\taj-2.jpg").convert('RGB')
img_np = np.array(img_color)

# Convert to grayscale
gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)

# Detect corners using Shi-Tomasi
corners = cv2.goodFeaturesToTrack(gray, maxCorners=200, qualityLevel=0.01, minDistance=10)

# Convert to integer coordinates
corners = np.int0(corners)

# Draw corners on the image
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img_np, (x, y), 4, (255, 0, 0), -1)

# Show result
plt.figure(figsize=(10, 8))
plt.imshow(img_np)
plt.title("Shi-Tomasi Corner Detection")
plt.axis("off")
plt.show()
```
